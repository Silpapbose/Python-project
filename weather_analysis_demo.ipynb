# 🌦️ Weather Analyzer - Demo

This notebook demonstrates how to use the Weather Analyzer Python package to analyze weather data.

Import libraries

```python
import pandas as pd
from weather_analyzer.loader import load_data
from weather_analyzer.analysis import analyze_data
```

Load dataset

```python
df = load_data("data/weather.csv")
df.head()
```

Basic info

```python
df.info()
```

Run your analyzer

```python
analyze_data(df, "Temperature (C)")
```
